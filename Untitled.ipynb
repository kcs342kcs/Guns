```python
#setup and dependences

import pandas as pd
import os
import numpy as np
```

```python
#files to load
guns_data= ("guns_output.csv")

#readfile
guns_file= pd.read_csv(guns_data)

guns_file.head()
                    
```

```python
#create new df w/ dropped columns
#gapminder_ocean.drop(['pop', 'gdpPercap', 'continent'], axis=1)
all_permits_df= guns_file.drop(['permit_recheck',"multiple", "admin", "prepawn_handgun", "prepawn_long_gun",  
                "rentals_handgun", "rentals_long_gun", "private_sale_handgun", "private_sale_long_gun", 
                "private_sale_other", "return_to_seller_handgun", "return_to_seller_long_gun", "return_to_seller_other",
                'prepawn_other', 'redemption_handgun', 'redemption_long_gun', 'redemption_other', 'returned_handgun',
                'returned_long_gun', 'returned_other' ], axis=1)

all_permits_df
```

```python
#bin by year and asum permits for each year
permit_groups= all_permits_df.groupby("month")
permit_sums= permit_groups["permit", 'handgun', 'long_gun', 'other'].sum()
permit_sums
```
